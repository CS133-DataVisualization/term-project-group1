```python
# 시가총액 (Market cap)
# Market cap = stock price x Shares Outstanding(발행주식수)
# Extract Shares Outstanding
# Extract close price in stock csv files
# Calculates market cap = close x Shares Outstanding
# Merge df_rev and market cap table

import pandas as pd
import os
import re
import matplotlib.pyplot as plt

base_dir = "/content/drive/MyDrive/cs/cs133/dataset/apple_financial_data"

quarters = []
shares_list = []

for folder in sorted(os.listdir(base_dir)):
    m = re.match(r"FY(\d{2})_Q(\d)_Raw", folder)
    if not m:
        continue

    fiscal = m.group(1)
    quarter = m.group(2)

    csv_path = os.path.join(base_dir, folder, f"FY{fiscal}_Q{quarter}_IS_raw.csv")

    if not os.path.exists(csv_path):
        continue

    df = pd.read_csv(csv_path)
    col0 = df.columns[0]

    # 1) 'Shares used in computing earnings per share' 위치 찾기
    anchor_idx = df[df[col0].str.contains("Shares used in computing earnings per share", na=False)].index

    if len(anchor_idx) == 0:
        print(f"⚠ anchor not found in {csv_path}")
        continue

    anchor_idx = anchor_idx[0]

    # 2) anchor 아래 몇 줄 안에서 'Diluted' 찾기
    sub = df.iloc[anchor_idx+1 : anchor_idx+6]   # 아래 5줄 정도만 보면 됨
    row = sub[sub[col0].str.contains("Diluted", na=False)]

    if len(row) == 0:
        print(f"⚠ Diluted shares not found near anchor in {csv_path}")
        continue

    value = row.iloc[0, 1]   # 두 번째 column 값 (예: '14,994,082')
    value_str = str(value).replace(",", "").strip()

    # 보통 정수일 텐데, 혹시 모를 소수점 대비해서 float → int
    value_num = int(float(value_str)) * 1000   # 'thousands' → 실제 주식 수

    quarters.append(f"FY{fiscal}_Q{quarter}")
    shares_list.append(value_num)

df_shares = pd.DataFrame({
    "Quarter": quarters,
    "SharesOutstanding": shares_list
}).sort_values("Quarter").reset_index(drop=True)

df_shares
```

```python
# 3) Extract the last close price in quarter
stock_path = "/content/drive/MyDrive/cs/cs133/dataset/apple_financial_data/apple_stock_data_22_to_25.csv"
df_stock = pd.read_csv(stock_path)

df_stock["Date"] = pd.to_datetime(df_stock["Date"])
df_stock["Year"] = df_stock["Date"].dt.year
df_stock["Q"] = df_stock["Date"].dt.quarter

# 분기 마지막 거래일
df_stock_q = (
    df_stock.sort_values("Date")
            .groupby(["Year", "Q"])
            .tail(1)
            .reset_index(drop=True)
)

# FY22_Q1 같은 형태로 맞추기
df_stock_q["Quarter"] = df_stock_q.apply(
    lambda r: f"FY{str(r['Year'])[-2:]}_Q{int(r['Q'])}", axis=1
)

# Close 값을 숫자로 변환
df_stock_q["Close"] = (
    df_stock_q["Close"].astype(str).str.replace(",", "").astype(float)
)

df_stock_q = df_stock_q[["Quarter", "Close"]].sort_values("Quarter").reset_index(drop=True)
df_stock_q
```

```python
# 4) Caculates Market Cap
df_marketcap = pd.merge(df_stock_q, df_shares, on="Quarter", how="inner")

df_marketcap["MarketCap"] = df_marketcap["Close"] * df_marketcap["SharesOutstanding"]
df_marketcap["MarketCap_Trillion"] = df_marketcap["MarketCap"] / 1e12  # trillion USD

df_marketcap
```

```python
df_all = pd.merge(df_rev, df_marketcap, on="Quarter", how="inner")
df_all = df_all.sort_values("Quarter").reset_index(drop=True)
df_all
```

```python
# 6) 매출 vs 시가총액 그래프
plt.figure(figsize=(14,6))

ax1 = plt.gca()

# ① 매출 (왼쪽 y축)
ax1.plot(
    df_all["Quarter"],
    df_all["Revenue"],
    marker='o',
    linewidth=2,
    label="Revenue (Million USD)"
)
ax1.set_ylabel("Revenue (Million USD)")
ax1.set_xlabel("Quarter")

# ② 시가총액 (오른쪽 y축, Trillion 단위)
ax2 = ax1.twinx()
ax2.plot(
    df_all["Quarter"],
    df_all["MarketCap_Trillion"],
    marker='s',
    linewidth=2,
    color="green",
    label="Market Cap (Trillion USD)"
)
ax2.set_ylabel("Market Cap (Trillion USD)")

# x축 라벨 회전
plt.xticks(rotation=45)

# 그리드 & 타이틀
ax1.grid(axis='y', linestyle='--', alpha=0.4)
plt.title("Apple Quarterly Revenue vs Market Cap (FY22_Q1 ~ FY25_Q4)")

# 레전드 합치기
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# Find PS Ratio (Market Cap / Revenue)
# 시장 기대감을 수치화하는 지표
df_all["PS_Ratio"] = df_all["MarketCap"] / (df_all["Revenue"] * 1e6)
```

```python
# Generate features
df_all["Revenue_QoQ"] = df_all["Revenue"].pct_change()
df_all["Revenue_YoY"] = df_all["Revenue"].pct_change(4)
df_all["Revenue_Level"] = df_all["Revenue"]
df_all["Revenue_Volatility"] = df_all["Revenue"].rolling(4).std()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

features = ["Revenue_Level", "Revenue_QoQ", "Revenue_YoY", "Revenue_Volatility"]
X = df_all[features].fillna(0)
y = df_all["PS_Ratio"]

model = RandomForestRegressor(n_estimators=500, random_state=0)
model.fit(X, y)
```

```python
importances = model.feature_importances_
plt.bar(features, importances)
plt.title("Feature Importance for Predicting PS Ratio")
plt.show()
```

```python
"""
Feature Importance

Revenue YoY (전년 대비 성장률) → 55%

Revenue Level (매출 절대 규모) → 23%

Revenue QoQ → 11%

Revenue Volatility → 10%

“성장률(즉, 미래 기대)”이 매출 자체보다 훨씬 중요하다.
"""
# Calculate Revenue YoY
df_all["Revenue_YoY"] = df_all["Revenue"].pct_change(4)

# Min-Max Scaling for Close price
df_all["Close_norm"] = (df_all["Close"] - df_all["Close"].min()) / (df_all["Close"].max() - df_all["Close"].min())

plt.figure(figsize=(14,6))

# YoY 성장률
plt.plot(df_all["Quarter"], df_all["Revenue_YoY"], marker='o', label="Revenue YoY %")

# 정규화된 주가
plt.plot(df_all["Quarter"], df_all["Close_norm"], marker='s', label="Stock Price (Normalized)")

plt.title("Revenue YoY Growth vs Stock Price (Normalized)")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.show()
```

```python
# 해석
# Revenue YoY % (전년 대비 성장률), 분기별 데이터 값은 (해당 분기 vs 전년 같은 분기)로 계산
# 그래프 결과
# Revenue YoY의 상승 + 주가 하락 (다음 분기 실적 기대로 인해, 다음 분기 발표 전까지 주가 상승)
# 즉, "전년 대비 성장률"의 상승이 주가 상승을 선반영함.
```
