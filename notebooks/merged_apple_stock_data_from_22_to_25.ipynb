```python
import pandas as pd
import glob
import os

base_path = "/content/drive/MyDrive/cs/cs133/dataset/apple_financial_data/apple_stock_data"
file_list = glob.glob(base_path + "/*.csv")

# extract file names
file_list = sorted(file_list, key=lambda x: os.path.basename(x))
file_names = [os.path.basename(f) for f in file_list]

print("ðŸ“„ apple_stock_data/ CSV files:")
for name in file_names:
    print(" -", name)

print(f"\n total: {len(file_names)}CSV files.")
```

```python
# combine all csv files to one
df = pd.concat((pd.read_csv(file) for file in file_list), ignore_index=True)
```

```python
save_path = "/content/drive/MyDrive/cs/cs133/dataset/apple_financial_data/apple_stock_data_22_to_25.csv"

df.to_csv(save_path, index=False)
```
